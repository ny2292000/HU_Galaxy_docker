```python
import numpy as np
import matplotlib.pyplot as plt

def rotation_matrix(theta):
    c, s = np.cos(theta), np.sin(theta)
    return np.array([[c, -s], [s, c]])

def plotme(ax, theta, wavelength, label="", periods=3):
    # Convert theta from degrees to radians for np functions
    theta_rad = np.radians(theta+90)
    
    # Create a rotation matrix for the given angle
    rot_matrix = rotation_matrix(theta_rad)

    # Define the range and calculate the sinusoidal values
    x = np.linspace(0, periods * wavelength, 1000)
    y = np.sin(2 * np.pi * x / wavelength)

    # Find the crossing points on the x-axis (where y = 0)
    crossing_points = [(n * wavelength, 0) for n in range(periods + 1)]

    # Rotate the crossing points
    rotated_crossing_points = [np.dot(rot_matrix, point) for point in crossing_points]

    # Rotate the sinusoidal curve
    xy_original = np.vstack((x, y))
    xy_rotated = np.dot(rot_matrix, xy_original)

    # Plot the rotated sinusoidal curve
    ax.plot(xy_rotated[0], xy_rotated[1], label=label)
    
    # Plot the crossing points
    for point in rotated_crossing_points:
        ax.plot(point[0], point[1], 'ro')  # 'ro' is the marker style for red circles

    # Draw the segment along the same angle as the sinusoid
    segment_end = rotated_crossing_points[-1]
    ax.plot([0, segment_end[0]], [0, segment_end[1]], 'k-')  # 'k-' is the style for a black line

    # Return the rotated crossing points
    return rotated_crossing_points

def draw_lines(rotated_crossing_points_1, rotated_crossing_points_2, ax,  linestyle="--"):
    # Draw lines between vertical and inclined sinusoid crossing points
    for p1, p2 in zip(rotated_crossing_points_1, rotated_crossing_points_2):
        ax.plot([p1[0], p2[0]], [p1[1], p2[1]], 'orange', linestyle='--')  # Draw line with orange color

def createPlot(alpha, fig, ax):
    # Example usage:
    theta = 45 -alpha
    lambda_r = 2 * np.pi
    lambda_light = lambda_r / np.cos(np.radians(theta))
    lambda_light_x = lambda_light/np.sin(np.radians(theta))
    
    label=f'Transition Dipole Oscillation'
    rotated_crossing_points_vertical = plotme(ax, 0, lambda_r, label=label)  # Plot a sinusoid with a wavelength of 2pi
    
    label=f'Light Propagation in 4D'
    rotated_crossing_points_inclined = plotme(ax, theta, lambda_light, label=label)  # Plot a sinusoid at a 45-degree angle with adjusted wavelength
    draw_lines(rotated_crossing_points_vertical, rotated_crossing_points_inclined, ax, linestyle="--")
    
    label=f'Light Propagation as seen in 3D'
    rotated_crossing_points_horizontal = plotme(ax, 90, lambda_light_x, label=label)  # Plot a sinusoid at a 45-degree angle with adjusted wavelength
    draw_lines(rotated_crossing_points_horizontal, rotated_crossing_points_inclined, ax, linestyle=".")
    
    # Set plot labels and title
    ax.set_xlabel('3D-Distance')
    ax.set_ylabel('Radial Direction Distance')
    ax.set_title(f'Light Propagation in 4D at {theta} degrees')
    ax.legend(loc='upper left', bbox_to_anchor=(1, 1))
    ax.grid(True)
     # Set aspect of the plot to be equal
    ax.set_aspect('equal')
    ax.set_xlim([-50,1])
    ax.set_ylim([-1,20])




for alpha in np.arange(20, -1,-1):  # Corrected to decrement from 30 to 5
    theta = 45 - alpha 
    fig, ax = plt.subplots()
    createPlot(alpha, fig, ax)
    plt.savefig(f"./ChangingRedshift//LightPropagationin4D_{theta}degrees.png", bbox_inches='tight')
    plt.close(fig)  # Close the figure after saving to free memory

```

```python
# Adjusted Python code to create a polar plot without the center shift, add orthogonal axes, radii labels, and point labels

import numpy as np
from matplotlib.patches import Arc
import matplotlib.patches as patches
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate alpha_0 given r_inner and r_outer
def calculate_alpha_0(r_inner, r_outer):
    # Using the arcsin function to calculate alpha_0 based on r_inner and r_outer
    alpha_0 = np.arcsin((r_inner / r_outer) * (np.sqrt(2) / 2)) + np.pi/4
    return alpha_0

# Function to calculate the points A and B given r_inner, r_outer and alpha_0
def calculate_points_A_B(r_inner, r_outer, alpha_0):
    # Point A (on r_inner circle, angle alpha_0 from the horizontal axis)
    A_x = r_outer * np.cos(np.pi/2)
    A_y = r_outer * np.sin(np.pi/2)

    # Point B (on r_outer circle, angle pi/4 from the horizontal axis)
    B_x = r_inner * np.cos(np.pi/2 - alpha_0)
    B_y = r_inner * np.sin(np.pi/2 - alpha_0)

    return (A_x, A_y), (B_x, B_y)

# Function to calculate alpha_1 and Gamma for each intermediate radius
def calculate_alpha_1_Gamma(r_inner, r_outer, n):
    # Calculate alpha_0
    alpha_0 = calculate_alpha_0(r_inner, r_outer)

    # Calculate Points A and B
    point_A, point_B = calculate_points_A_B(r_inner, r_outer, alpha_0)

    # Initialize lists to store the results
    alphas_1 = []
    Gammas = []

    # Create n radii between r_inner and r_outer and calculate alpha_1 and Gamma for each
    radii = np.linspace(r_inner, r_outer, n)
    for r in radii:
        # Calculate alpha_1 using the Law of Sines
        A = r_outer/r
        B = np.tan(alpha_0-np.pi/4)
        AA = 1+B**2
        BB = -2*A*B**2
        CC = B**2*A**2-1
        sol = (-BB+np.sqrt(BB**2-4*AA*CC))/(2*AA)
        alpha_1 = np.pi/2 - np.arccos(sol)
        # Calculate Gamma
        Gamma = (np.pi / 4) + alpha_0 - alpha_1

        # Store the results in the lists
        alphas_1.append(alpha_1)
        Gammas.append(Gamma)
    alphas_1_ = np.array(alphas_1)
    Gammas_ = np.array(Gammas)
    return alphas_1_, Gammas_, radii, point_A, point_B

def plotArc(alpha0, r0, label, ax):
    ax.plot(np.linspace(0, alpha0, 100), [r0]*100, color='red', linestyle='-', linewidth=0.5)  # Polar grid lines
    ax.text(alpha0/2, r0*0.9, label, fontsize=14, ha='center', va='center')
    return ax


# Setup the figure and axis
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'projection': 'polar'})

#############################################################################
#############################################################################
#############################################################################
radius_inner = 6  # Example value for inner radius
radius_intermediate = 10
radius_outer = 14  # Example value for outer radius
n =  10000  # Number of radii to calculate
#############################################################################
#############################################################################
#############################################################################
r = np.linspace(radius_inner, radius_outer, n)
alpha = np.pi/2 - (radius_outer - r)/radius_outer
ax.plot(alpha,r)

# Add circles at the specified radii
circle_inner = plt.Circle((0, 0), radius_inner, transform=ax.transData._b, color='green', 
                          linewidth=2, fill=False)
circle_outer = plt.Circle((0, 0), radius_outer, transform=ax.transData._b, color='red', 
                          linewidth=2, fill=False)

# Add the circles to the plot
ax.add_artist(circle_inner)
ax.add_artist(circle_outer)

# Plot segments at specified angles
angle_90_deg = np.deg2rad(90)
angle_alpha_0 = np.pi/2 - calcAlpha0(radius_inner, radius_outer) 
angle_offset =np.pi/2 - radius_inner / radius_outer  # 6/14 radians offset
angle_offset_1 = np.pi/2 - (radius_outer - radius_intermediate) / radius_outer  # 4/14 radians offset
angle_1_rad = np.pi/2 -  1  # One radian

# Draw segments
ax.plot([angle_90_deg, angle_90_deg], [0, radius_outer], color='black', linewidth=1)
ax.plot([ angle_alpha_0, angle_alpha_0], [0, radius_outer], color='black', linewidth=1)
ax.plot([angle_offset_1,angle_offset_1], [0, radius_outer], color='green', linewidth=1, linestyle="--")
ax.plot([angle_1_rad, angle_1_rad], [0, radius_outer], color='black', linewidth=1)
ax.plot([angle_alpha_0, angle_90_deg], [radius_inner, radius_outer], color='black', linewidth=1)

ax=plotArc(angle_alpha_0, 0.8*radius_inner,r'$\alpha_0$', ax)
ax=plotArc(angle_offset_1, 0.4*radius_inner, r'$\alpha_1$', ax)

# Mark the end points A and B
ax.plot(np.pi/2, radius_outer, 'o', label='Point A (outer radius)', color='green')
ax.plot(angle_alpha_0, radius_inner, 'o', label='Point B (inner radius)', color='red')
r_alpha_1_grid = np.linspace(radius_outer, radius_inner, 1000)
alpha_1_grid = np.arcsin(radius_inner/r_alpha_1_grid*np.sin(3*np.pi/4))-np.pi/4+angle_alpha_0

# Add orthogonal axes
ax.plot([0, 0], [0, radius_outer], color='blue', linestyle='-', linewidth=1.5)

# Add the polar grid for even radii only
radii_grid = np.arange(2, radius_outer+1, 2)  # Radii for the grid lines
for r in radii_grid:
    ax.plot(np.linspace(0, 2*np.pi, 100), [r]*100, color='grey', linestyle='--', linewidth=0.5)  # Polar grid lines
    ax.text(0, r, str(r), fontsize=10, ha='center', va='bottom')  # Radii labels

# Label points A, B, C, and O
labels = ['B', "C", 'A',  '1 rad', 'O']
angles = [  -0.035 + angle_alpha_0, 
           angle_offset - 0.04, 
          angle_90_deg + 0.04, 
          angle_1_rad -0.03, 
          0]
r_shift = 1.05
radii = [radius_inner*r_shift, 
         radius_outer*r_shift, 
         radius_outer*r_shift*0.91, 
         radius_outer*r_shift*1.02, 
         0]
for label, angle, radius in zip(labels, angles, radii):
    ax.text(angle, radius, label, fontsize=14, ha='center', va='center')


# Annotate the angle (Gamma)
ax.annotate(r'$\Gamma$', 
             xy=(0, 0), 
             xytext=(30, 155), 
             textcoords='offset points',
             fontsize=12)

# Annotate the angle (pi/4)
ax.annotate(r'$\frac{\pi}{4}$', 
             xy=(0, 0), 
             xytext=(40, 100), 
             textcoords='offset points',
             fontsize=12)

# Annotate the angle (3pi/4)
ax.annotate(r'$\frac{3\pi}{4}$', 
             xy=(0, 0), 
             xytext=(25, 70), 
             textcoords='offset points',
             fontsize=12)


# # Adding the arcs for angles alpha_0, alpha_1, and Gamma without altering previous elements

arc = patches.Arc((0, 9),  # Center point
                  4,       # Width
                  4,       # Height
                  theta1=43,        # Starting angle
                  theta2=50,  # Ending angle
                  color='green',
                  linewidth=1)
ax.add_patch(arc)




# Set the display parameters for the polar plot
ax.set_title("HU View of the Past", va='bottom')
ax.set_ylim(0, radius_outer*1.05)  # setting limit for radial direction
ax.set_xticks([])  # Remove angular ticks
ax.set_yticks([])  # Remove radial ticks
ax.spines['polar'].set_visible(False)  # Remove the polar spine
plt.savefig(f"./HU_View_The_Past.png", bbox_inches='tight')
# Display the plot
plt.show()
```
