```python
from pptx import Presentation

# Function to extract text and notes from a presentation
def extract_text_from_pptx(file_path):
    prs = Presentation(file_path)
    text_content = []
    for slide in prs.slides:
        slide_content = {"slide_number": slide.slide_id, "text": [], "notes": ""}
        for shape in slide.shapes:
            if hasattr(shape, "text"):
                slide_content["text"].append(shape.text)
        if slide.has_notes_slide:
            notes_slide = slide.notes_slide
            notes_text = notes_slide.notes_text_frame.text
            slide_content["notes"] = notes_text
        text_content.append(slide_content)
    return text_content

# Define paths to the presentations
file3_path = "/home/mp74207/Downloads/HU_BigPop.pptx"
file4_path = "/home/mp74207/Downloads/Big_Pop_Cosmogenesis.pptx"
combined_pptx_path = "/home/mp74207/Downloads/Combined_Big_Pop_Cosmogenesis.pptx"

# Extract text from the new presentations
file3_content = extract_text_from_pptx(file3_path)
file4_content = extract_text_from_pptx(file4_path)

# Create a new presentation and add the combined content
combined_prs = Presentation()

def add_slide(prs, slide_content):
    slide_layout = prs.slide_layouts[1]  # Use a layout with title and content
    slide = prs.slides.add_slide(slide_layout)
    for idx, text in enumerate(slide_content['text']):
        if idx == 0:
            title = slide.shapes.title
            title.text = text
        else:
            content = slide.placeholders[1]
            text_frame = content.text_frame
            p = text_frame.add_paragraph()
            p.text = text
    if slide_content['notes']:
        notes_slide = slide.notes_slide
        notes_text_frame = notes_slide.notes_text_frame
        notes_text_frame.text = slide_content['notes']

# Add slides from both presentations
for slide_content in file3_content:
    add_slide(combined_prs, slide_content)

for slide_content in file4_content:
    add_slide(combined_prs, slide_content)

# Save the combined presentation
combined_prs.save(combined_pptx_path)

combined_pptx_path
```
