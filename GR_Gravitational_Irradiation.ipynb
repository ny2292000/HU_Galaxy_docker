To derive the formula for the emission of gravitational waves via the quadrupole moment and then specifically for an elliptical orbit in a binary system, we need to delve into the theory of general relativity's quadrupole formula for gravitational radiation. Here, I'll outline the steps and concepts involved in these derivations.

### Step 1: Quadrupole Moment Formula

The quadrupole moment $ Q_{ij} $ of a mass distribution, which is a measure of the system's shape relative to spherical symmetry, is defined as:
$
Q_{ij} = \int \rho(\vec{r}) (3x_i x_j - r^2 \delta_{ij}) d^3 r
$
where $ \rho(\vec{r}) $ is the mass density, $ x_i, x_j $ are the Cartesian coordinates, $ r $ is the distance from the center of mass to the point in the mass distribution, and $ \delta_{ij} $ is the Kronecker delta.

### Step 2: Gravitational Wave Power from Quadrupole Moment

The power emitted in the form of gravitational waves by a system with a changing quadrupole moment can be approximated by the quadrupole formula:
$
P = \frac{G}{5c^5} \langle \dddot{Q}_{ij} \dddot{Q}^{ij} \rangle
$
where $ \dddot{Q}_{ij} $ represents the third time derivative of the quadrupole moment tensor $ Q_{ij} $, and the angle brackets denote averaging over several wavelengths or periods of the wave.

### Step 3: Quadrupole Moment for a Binary System

For a binary system, especially in an elliptical orbit, the quadrupole moment can be particularly complex due to the non-circular motion. However, we can model the system assuming the masses $ m_1 $ and $ m_2 $ orbit their common center of mass in elliptical paths defined by:
$
r(t) = \frac{a(1 - e^2)}{1 + e \cos \theta(t)}
$
where $ a $ is the semi-major axis, $ e $ is the eccentricity, $ \theta(t) $ is the true anomaly, and $ r(t) $ is the distance between the two masses at time $ t $.

### Step 4: Computing the Quadrupole Moment for Elliptical Orbits

The expressions for $ x_i $ and $ x_j $ components will depend on the orbital parameters and positions of $ m_1 $ and $ m_2 $. For simplicity, assume the orbit lies in the $ xy $-plane:
$
Q_{xx} = \int \rho(\vec{r}) (3x^2 - r^2) d^3 r, \quad Q_{yy} = \int \rho(\vec{r}) (3y^2 - r^2) d^3 r, \quad \text{etc.}
$
For the binary system, if we approximate each mass as a point mass, then:
$
Q_{xx} = m_1 (3x_1^2 - r_1^2) + m_2 (3x_2^2 - r_2^2)
$
And similar for other components, considering $ x_1, y_1, x_2, y_2 $ are the coordinates of $ m_1 $ and $ m_2 $.

### Step 5: Time Derivatives of Quadrupole Moments

The next step is to calculate the third time derivative of these components, $ \dddot{Q}_{ij} $, which involves differentiating the positions and velocities of the masses with respect to time. These derivatives depend on the velocities and accelerations, which in turn depend on the gravitational interaction between the masses and their motion along the elliptical path.

````python
from astropy import units as uu, constants as cc
import numpy as np

# To compare Electrostatic and Gravitation one should use the same amount of 4D mass on both.
# on Gravitation 4D Mass of 1kg is equal to a 3D Mass of 1 Kg
mH = (cc.m_p + cc.m_e).si

F_gravitation = cc.G*(1*uu.kg)**2/(1*uu.m)**2

# For the Electrostatic Calculation the 4D mass of an Electron is 0.5 of mH
# so 1kg of electrons is uu.kg/(0.5*mH)*cc.e
Charge = (1*uu.kg/mH)*cc.e.si

F_electrostatic =1/(4*np.pi*cc.eps0)*Charge**2/(1*uu.m)**2
a = (F_gravitation/F_electrostatic).si
a
````

````python
Ng = 1/(cc.m_p + cc.m_e).si.value
Ne = 2*Ng


Proposed_Ratio_GE = 36* cc.G.si.value/(4*np.pi*cc.eps0.si.value)* Ng/Ne
Proposed_Ratio_GE
````
